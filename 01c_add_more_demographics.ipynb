```python
import pandas as pd

# Set pandas display rows to 500
pd.set_option('display.max_rows', 500)
```

```python
data = pd.read_csv('./data/collated_data_amb.csv')
data.head().T
```

```python
census = pd.read_csv('./data/demographic/2011_Census_Data.csv')
census.head().T
```

```python
cols_to_merge = [
    'SOA Name',
    'Supergroup Code',
    'Supergroup Name',
    'Group Name',
    'Group Code',
    '% Persons living in a communal establishment',
    'Number of persons per heactre',
    '% Persons aged 0 to 4 years',
    '% Persons aged 5 to 14 years',
    '% Persons aged 25 to 44 years',
    '% Persons aged 45 to 64 years',
    '% Persons aged 65 to 89',
    '% Persons aged 90 years and over',
    '% Persons aged 16 and over who are single',
    '% Persons aged 16 and over who are married or registered in a civil partnership',
    '% Persons aged 16 and over who are divorced or separated',
    '% Persons who are white',
    '% Households who live in a detached house or bungalow',
    '% Households who live in a semi-detached house or bungalow',
    '% Households who live in a terrace house or end of terrace house',
    '% Households who live in a flat',
    '% Households who own or have shared ownership of property',
    '% Households who are private renting',
    '% Households who are social renting',
    'Standard Illness ratio',
    '% Persons providing unpaid care',
    '% Persons aged 16 and over whose highest level of qualification is Level 1, Level 2 or apprenticeship',
    '% Persons aged 16 and over whose highest level of qualification is Level 3 qualifications',
    '% Persons aged 16 and over whose highest level of qualification is Level 4 and above',
    '% Persons aged 16 and over who are schoolchildren or full-time students',
    '% Households with two or more cars',
    '% Persons aged between 16 and 74 who use public transport to get to work',
    '% Persons aged between 16 and 74 who use private transport transport to get to work',
    '% Persons aged between 16 and 74 who walk, cycle or use an alternative method of transport to get to work',
    '% Employed persons aged between 16 and 74 who are unemployed',
    '% Employed persons aged between 16 and 74 who are part-time',
    '% Employed persons aged between 16 and 74 who are full-time',
]

merged_data = data.merge(census[cols_to_merge], left_on='LSOA', right_on='SOA Name', how='left')
# remove soa name
merged_data.drop('SOA Name', axis=1, inplace=True)
```

```python
merged_data.head()
merged_data.to_csv('./data/collated_data_amb_census.csv', index=False)
```
